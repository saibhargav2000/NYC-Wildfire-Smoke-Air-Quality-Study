# Air Quality Graphs


```
import pandas as pd
import matplotlib.pyplot as plt
from scipy.stats import shapiro, levene, ttest_ind
```

```
data = pd.DataFrame({
 'time':['pre','during','post'],
 'inside':[0.15,2.06,-0.18],
 'outside':[0.14,2.03,-0.21]
})
data
```

```
plt.plot(data['time'],data['inside'],marker='o',label='inside')
plt.plot(data['time'],data['outside'],marker='o',label='outside')
plt.legend()
plt.title('Air Pollution Comparison')
plt.show()
```
